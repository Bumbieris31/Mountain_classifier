```python
from fastai.vision import *
```

```python
folder = 'alps'
file = 'alps.txt'
```

```python
folder = 'rocky_mountains'
file = 'rocky_mountains.txt'
```

```python
folder = 'himalayas'
file = 'himalayas.txt'
```

```python
path = Path('data/mountains')
dest = path/folder
dest.mkdir(parents=True, exist_ok=True)
```

```python
path.ls()
```

```python
##Downloading images
```

```python
classes = ['alps', 'rocky_mountains', 'himalayas']
download_images(path/folder/file, dest, max_pics=200)
```

```python
for c in classes:
    print(c)
    verify_images(path/c, delete=True, max_size=500)
```

```python
###Randomly assigning 20% of the dataset for training.
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.2,
        ds_tfms=get_transforms(), size=224, num_workers=4).normalize(imagenet_stats)
```

```python
data.classes
```

```python
data.show_batch(rows=3, figsize=(7,8))
```

```python
data.classes, data.c, len(data.train_ds), len(data.valid_ds)
```

```python
#Training
```

```python
learn = cnn_learner(data, models.resnet34, metrics=error_rate)
```

```python
learn.fit_one_cycle(4)
```

```python
learn.save('stage-1')
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(2, max_lr=slice(1e-04, 1e-025))
```

```python
learn.save('stage-2')
```

```python
##Interpretation
```

```python
learn.load('stage-2')
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_confusion_matrix()
```
